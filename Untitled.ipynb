```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Romania'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
tabel = soup.find_all('table')[0]
```

```python
titluri = tabel.find_all('th')
```

```python
titluri_tabel = [titlu for titlu in titluri]
print(titluri_tabel)
```

```python
titluri_tabel = [titlu.text.strip() for titlu in titluri]

print(titluri_tabel)
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(columns = titluri_tabel)
data
```

```python
column_data= tabel.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data =  row.find_all('td')
    titluri_randuri = [date.text.strip() for date in row_data]
    lungime = len(data) #lungimea frameului(tabelului)
    
    data.loc[lungime] = titluri_randuri
```
